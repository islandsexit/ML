### Выделение главных атрибутов и представление их в удобном для работы виде

В данном `Dataframe` можем заметить, что у нас есть 5 главных атрибутов:
- Название компетенции(1 колонка)
- Описание компетенции(2 колонка)
- Раздел компетенции(3 колонка)
- Значимость раздела(4 колонка)
- Признаки раздела(5 колонка)

Последние 3 признака являются шаблоном для остальных колонок в `Dataframe`

Создадим новый `Dataframe`, где в конечном счете, у нас останутся предобработанные данные

```python
df = pd.DataFrame({'название_компетенции':[],
                  'описание_компетенции':[],
                  'название_раздела':[],
                  'значимость_раздела':[],
                  'признаки_раздела':[]})

for i in range(27):
    df['col_'+str(i)] = np.nan
  
#приведем колонки Dataframe к типу string за счет чего сможем избежать ошибки при препроцессинге
for data_types in df.columns:
    df[data_types] = df[data_types].astype(str)
```

### Формирование структуры данных 

В вышеописанном мы смогли выделить главные атрибуты нашего датасета \
Поэтому в предобработке будем формировать именну такую структуру данных \
А именно 5 колонок с названием компетенции и информацией о ее разделах

```python
#проходим по каждой строчке исходного набора дыннх
for row in range(len(data)):
    #выделяем шаблонизированные наборы данных 
    for a,b,c in zip(range(2,28,3), range(3,28,3), range(4,28,3)):
        #логическая операция для прерывание нашего цикла при остутвии входных данных
        if pd.isnull(data.iloc[row][a]) == True:
            break
        else:
            #экспорт формированой структуры данных из хаотично распределенных в исходных данных
            df.loc[len(df)] = np.nan
            df = df.shift()
            df.loc[0,'название_компетенции'] = data.iloc[row][0]
            df.loc[0,'описание_компетенции'] = data.iloc[row][1]
            df.loc[0,'название_раздела'] = data.iloc[row][a]
            df.loc[0,'значимость_раздела'] = data.iloc[row][b]
            df.loc[0,'признаки_раздела'] = data.iloc[row][c]
            
#удаление сущностей не несущих логического смысла из предобработанного DataFrame
df.drop(df.iloc[:,5:], inplace=True, axis = 1)
```

### Предобработка пустых значений

Посмотрим распределение плотности в наших данных

```python
df.dropna(inplace=True)
```

Проведем заново распределение `index` нашего `DataFrame`

```python
df = df.reset_index(drop=True)
```

В методе `object.info()` можно увидеть, что `значимость_раздела` имеет формат строки, для дальнейшей работы приведем его к формату `float`, используя функцию `pd.to_numeric` с удалением неподходящих для форматирования в число значений (параметр ошибки `coerce`)

и заполним эти данные математической медианой, чтобы не потерять часть данных 

```python
df['значимость_раздела'] = pd.to_numeric(df['значимость_раздела'], errors='coerce')
df.fillna(df['значимость_раздела'].median(), inplace=True)
```

## Препроцессинг текстовых данных

### Очистка данных от шумовых строковых единиц

```python
#имортируем библиотеку nltk в которой есть собранные часто встречаемые слова, не несущих логическго смысла
import nltk

def clean_data(df):
    #Приыедем данные к нижнему регистру
    df = df.str.lower()
    #Регулярным выражением избавимся от специальных символов и от цифр 
    df = df.str.replace(r'[^а-яА-Яa-zA-Z]',' ', regex=True)
    #Избавимся от стоп-слов (слов, которые не несут смысла для модели)
    df = df.str.strip()
    df = df.apply(lambda x: ' '.join([word for word in x.split()
                                      if word not in nltk.corpus.stopwords.words('russian')]))
    return df
```

```python
df['описание_компетенции'] = clean_data(df['описание_компетенции'])
df['признаки_раздела'] = clean_data(df['признаки_раздела'])
df['название_раздела'] = clean_data(df['название_раздела'])
```

### Выделение новых атрибутов

Для снижения уровня ошибки выделим новые атрибуты, а именно воспользуемся математической суммой букв 
$f(x) = \sum_{text = 0}^{text}word$ \
Где
- `text` - это одномерный массив букв в тексте, где для каждой буквы своя позиция от 0 до конца текста
-`word` - представлена буквой в слове с булевым весом
-`f(x)` - математическая сумма количества слов в тексте

```python
#создадим пустые колонки для значений суммы
    # c_title - количество слов в названии раздела
df['c_title'] = np.nan
    # c_descr - количество слов в описании компетенции
df['c_descr'] = np.nan
    # c_title_feat - количество слов в признаках компетенции
df['c_title_feat'] = np.nan

# цикл для прохождения по всем строкам данных
for row in range(len(df)):
    df['c_title'][row] = int(len(df['название_раздела'][row].split()))
    df['c_descr'][row] = int(len(df['описание_компетенции'][row].split()))
    df['c_title_feat'][row] = int(len(df['признаки_раздела'][row].split()))
    
#При выделении новых атрибутов колонка сохраняется в float формате
#Но нецелочисленного выражение этих атрибутов быть не может
df['c_descr'] = df['c_descr'].astype(int)
df['c_title'] = df['c_title'].astype(int)
df['c_title_feat'] = df['c_title_feat'].astype(int)
```

### Отчет по первому этапу предобработки

Избавляясь от пустых значений, мы пришли к 100% плотности их распределения \
И данные готовы для дальнейшей работы

## Формирование словарей данных

В данном этапе будет рассматриваться Natural Texting Analize     
Создадим частотные словари на основе подхода `CountVectorizer` \
В который добавим функцию токенизации и стеммизации изобретенным Мартином Портером, а точнее его проект `SnowBallStemmer`, который является измененным стеммером Портера, но для индоевропейских языков

```python
#SnowballStemmer находится в библиотеке nltk, импортируем ее
from nltk.stem import SnowballStemmer
#импортируем re для регулярных выраженний, а именно для разделения слов на список слов
import re

stem = SnowballStemmer('russian')

def tokenization(text):
    text = re.split('\W+', text)
    text =[stem.stem(word) for word in text]
    return text
```

На данный момент мы создали токенизатор со стеммингом для нашего последующего формирования и анализа словаря данных 

```python
from sklearn.feature_extraction.text import CountVectorizer as CV
import csv
#min_df - удаляет слова, которые встречаются меньше чем в 10% документов
#max_df - удаляет слова, которые есть в 50% документах
cv = CV(stop_words= nltk.corpus.stopwords.words('russian'),
        tokenizer=tokenization, min_df=0.1, max_df=0.5)

vect_dict = cv.fit_transform(df['признаки_раздела'])
count_dict = cv.vocabulary_
words = cv.get_feature_names()

#сохраняем полученный частотный словарь данных
with open('dict.csv', 'w', encoding='utf-16', newline='') as f:
    w = csv.DictWriter(f,count_dict.keys())
    w.writeheader()
    w.writerow(count_dict)
```

```python
#в представленном виде находится словарь
pd.read_csv('dict.csv', encoding='utf-16')
```

## Анализ словарей 

Проанализируем возможность определения компетенции по атрибуту `признаки_раздела`
Для премера посмотрим возможность распределения на 4 кластера \
По вхождениям слов в векторайзер

```python
from sklearn.cluster import KMeans as KM

km = KM(n_clusters =4)

km.fit(vect_dict)
pred = km.predict(vect_dict)
```

На данном этапе, для понимания распределения кластеров, вызовем центроиды `Kmeans`, и их вхождения в векторайзере

```python
centers = km.cluster_centers_.argsort()[:,-1:-20:-1]
for num, est in enumerate(centers):
    print('\n'+str(num) +':\n'+', '.join(words[word] for word in est))
```

### Наименование кластеров
При наименовании кластеров может возникнуть проблема лишних слов, которые могут не нести никакого смысла. Тогда стоит ориентироватся на то, что есть

На основе выданных результатов, можно сказать, что:
* Кластер `0` можно назвать как "Информационные и коммуникационные технологии" из-за таких слов, как: систем, решен, программн, разработк, технолог, информац.
* Кластер `1` наменование этого кластера может быть проблематично, но т.к. здесь мало уникальных слов. Но тем не менее определим его, как "Строительство и строительные технологии" из-за слов: ситуац, техник, средств, норм, применен
* Кластер `2` можно отнести к "Сфера услуг" из за таких слов, как: труд, охра, прав, защит
* Кластер `3` можно отнести к "Производство и инженерные технологии" из-за таких слов, как: процесс, технолог, инструмент, техник, чертеж, эффективн, производств

А если у нас это получилось, значит наши данные имеют хорошую способность делиться на кластеры по содержанию своих значений
